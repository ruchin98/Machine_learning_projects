```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn import preprocessing, cross_validation, svm
from sklearn.linear_model import LogisticRegression,LinearRegression
```

```python
df1 = pd.read_csv('campaign_data.csv')
```

```python
df1['email_body'].head()
```

```python
del df1['subject']
df1.head()
```

```python
del df1['email_body']
df1.head()
```

```python
del df1['email_url']
df1.head()
```

```python
df12 = pd.get_dummies(df1)
```

```python
df12.head()
```

```python
df2 = pd.read_csv('train.csv')
```

```python
df2.head()
```

```python
df3 = pd.merge(df12,df2)
```

```python
df3.tail()
```

```python
 del df3['id']
del df3['user_id']
del df3['send_date']
```

```python
df3.head()
```

```python
X = np.array(df3.drop(['is_click','is_open'], 1))
y = np.array(df3['is_click'])
```

```python
X = preprocessing.scale(X)
```

```python
X_train, X_test, y_train, y_test = cross_validation.train_test_split(X, y, test_size=0.2)
```

```python
clf = LogisticRegression()
```

```python
clf.fit(X_train,y_train)
```

```python
conf = clf.score(X_test,y_test)
```

```python
conf
```

```python
df5 = pd.merge(df12,df4)
X = np.array(df5.drop(['id','user_id','send_date'], 1))
ans = clf.predict(X)
ans
df5['ans'] = ans
```
